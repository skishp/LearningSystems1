# Topic 5: Unsupervised Learning | kmeans algorithm 
In this file we will implement the k means algorithm.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans


X = np.load('OldFaithful.npy')

k = 2
kmeans = KMeans(k, random_state=0)
prediction = kmeans.fit(X).predict(X)

plt.scatter(X[:, 0], X[:, 1], c=prediction, s=40, cmap='viridis');
plt.show()
```

If you compare this graph to the one of the `gaussEM.ipynb` you will notice a difference in classification at x = 3.5 y = 65-70
