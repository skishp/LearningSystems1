# Topic: Naive Bayes
In this coding example, we will be implementing the naive bayes. For this we will use the provided Data out of the spamdata.mat file. 

```python
from sklearn.naive_bayes import *
from sklearn.model_selection import train_test_split
import scipy.io
from sklearn.metrics import *
import numpy as np
from matplotlib import pyplot as plt

def load_data():
    mat = scipy.io.loadmat('spamdata.mat')
    print(mat)
    x_train = mat["data"]
    y_train = mat["label"].T[:, 0]
    x_test = mat["newdata"]
    y_test = mat["newlabel"].T[:, 0]
    return x_train, x_test, y_train, y_test

x_train, x_test, y_train, y_test = load_data()

f, ax = plt.subplots(1,2)
hist_train = np.sum(x_train, axis=0) / np.sum(x_train)
hist_test = np.sum(x_test, axis=0) / np.sum(x_test)
xbar = np.arange(0, len(hist_train), 1)
ax[0].bar(xbar, hist_train)
ax[0].set_title("Train Data")
ax[1].bar(xbar, hist_test)
ax[0].set_ylim([0, 0.33])
ax[1].set_ylim([0, 0.33])
ax[1].set_title("Test Data")#
ax[1].set_xlabel("Feature")
ax[0].set_xlabel("Feature")
ax[0].set_ylabel("Frequency")
ax[1].set_ylabel("Frequency")


plt.show()
```

Now we can define our model and so on. 

```python
nb = BernoulliNB()
nb.fit(x_train, y_train)
y = nb.predict(x_test)
print(y)
print(x_test)
print(len(x_test))
print(len(y))
```
